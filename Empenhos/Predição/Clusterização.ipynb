```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
empenhos = pd.read_csv("../AnalisedeDados/empenhos-clean.csv", low_memory=False)
empenhos.head()
```

```python
df = empenhos.copy()
df.info()
```

```python
df.describe()
```

```python
# Plotando os dados originais
plt.figure(figsize=(11,7))
plt.scatter(df['codigo_municipio'],df['valor_empenhado'],c='gray')
plt.xlabel('Códigos Municipais')
plt.ylabel('Valores Empenhados')
plt.title('Dados Originais')
plt.show()
```

```python
# standardizing the data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
data_scaled = scaler.fit_transform(df)

# statistics of scaled data
pd.DataFrame(data_scaled).describe()
df_scaled = pd.DataFrame(data_scaled)
df_scaled
```

```python
# Plotando os dados padronizados
CODIGO_MUNICIPIO = 0
VALOR_EMPENHADO = 8
plt.figure(figsize=(11,7))
plt.scatter(df_scaled[CODIGO_MUNICIPIO],df_scaled[VALOR_EMPENHADO],c='gray')
plt.xlabel('Códigos Municipais')
plt.ylabel('Valores Empenhados')
plt.title('Dados Escalados')
plt.show()
```

```python
# Aplicando algoritmo K-means
from sklearn.cluster import KMeans
from sklearn.metrics import davies_bouldin_score
```

```python
#Encontrar o número de K de acordo com Davies-Bouldin
# com dataset padronizado
scores = []
centroids = list(range(2,8))
for centroid in centroids:
    kmeans = KMeans(n_clusters=centroid, init='k-means++')
    model = kmeans.fit_predict(df_scaled)
    
    # Calculate Davies Bouldin score
    score = davies_bouldin_score(df_scaled, model)
    
    scores.append(score)
    
plt.plot(centroids, scores, linestyle='--', marker='o', color='b');
plt.xlabel('K');
plt.ylabel('Davies Bouldin score');
plt.title('Davies Bouldin score vs. K');
```

```python
def kmeans_(X, n_cluster):
    km = KMeans(n_clusters=n_cluster, init='k-means++')

    y_km = km.fit_predict(X)
    print(y_km)
    
    for i in range(n_cluster):
        plt.scatter(X[y_km == i,CODIGO_MUNICIPIO], X[y_km == i,VALOR_EMPENHADO],label='cluster '+str(i))
    
    # plot the centroids
    plt.scatter(
        km.cluster_centers_[:, CODIGO_MUNICIPIO], km.cluster_centers_[:, VALOR_EMPENHADO],
        s=250, marker='*',
        c='red', edgecolor='black',
        label='centroids'
    )
    plt.legend(scatterpoints=1)
    plt.grid()
    plt.show()
```

```python
for i in range(2,9):
    kmeans_(df_scaled.to_numpy(), i)
```
