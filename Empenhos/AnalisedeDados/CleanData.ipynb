```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
empenhos = pd.read_csv("empenhos-pre.csv", low_memory=False)
```

```python
empenhos.head()
```

```python
features = ['codigo_municipio', 'codigo_orgao', 'codigo_unidade', 'codigo_funcao', 'codigo_subfuncao',
           'codigo_programa', 'codigo_elemento_despesa','modalidade_empenho','valor_empenhado','tipo_processo_licitatorio',
            'estado_empenho','codigo_tipo_negociante','situacao_favorecido']
df = empenhos[features]
df
```

```python
df.info()
```

```python
df[df['situacao_favorecido'].isna()]
```

```python
processos_licitatorios = df['tipo_processo_licitatorio'].unique()
for i, value in enumerate(processos_licitatorios):    
    print(i, value)
    df.loc[(df['tipo_processo_licitatorio']==value) | (df['tipo_processo_licitatorio'].isnull()),'tipo_processo_licitatorio'] = i
```

```python
situation = df['situacao_favorecido'].unique()
for i, value in enumerate(situation):    
    print(i, value)
    df.loc[(df['situacao_favorecido']==value) | (df['situacao_favorecido'].isnull()),'situacao_favorecido'] = i
```

```python
situation = df['modalidade_empenho'].unique()
for i, value in enumerate(situation):    
    print(i, value)
    df.loc[(df['modalidade_empenho']==value) | (df['modalidade_empenho'].isnull()),'modalidade_empenho'] = i
```

```python
situation = df['estado_empenho'].unique()
for i, value in enumerate(situation):    
    print(i, value)
    df.loc[(df['estado_empenho']==value) | (df['estado_empenho'].isnull()),'estado_empenho'] = i
```

```python
situation = df['codigo_elemento_despesa'].unique()
for i, value in enumerate(situation):    
    print(i, value)
    df.loc[(df['codigo_elemento_despesa']==value) | (df['codigo_elemento_despesa'].isnull()),'codigo_elemento_despesa'] = i
```

```python
feats = features
feats.remove("valor_empenhado") 
```

```python
for f in feats:
    print(f)
    df[f] = df[f].astype("category")
```

```python
df.info()
```

```python
df.to_csv('empenhos-clean.csv',index=False)
```
